```python
import tensorflow as tf
import keras
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
from tensorflow.keras import regularizers
0.00005
```

```python
model= Sequential([
    Dense(units=10, activation='relu',kernel_regularizer=regularizers.L2(0.00005)),
    Dense(units=5,activation='relu',kernel_regularizer=regularizers.L2(0.00005)),
    Dense(units=1,activation='linear',kernel_regularizer=regularizers.L2(0.00005))
])
```

```python
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.optimizers import Adam
model.compile(loss=MeanSquaredError(),optimizer=Adam(0.001))
```

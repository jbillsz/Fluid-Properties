```python
import tensorflow as tf
import keras
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
from tensorflow.keras import regularizers
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.optimizers import Adam
import pandas as pd
import numpy as np
raw_data=pd.read_csv('C:/Users/Paa Kwasi/Desktop/Density.csv')
raw_data
```

```python
model= Sequential([
    Dense(units=15, activation='relu',kernel_regularizer=regularizers.L2(0.000001)),
    Dense(units=10,activation='relu',kernel_regularizer=regularizers.L2(0.000001)),
    Dense(units=1,activation='linear',kernel_regularizer=regularizers.L2(0.000001))
])
```

```python
model.compile(loss=MeanSquaredError(),optimizer=Adam(0.001),metrics=['mean_squared_error','mean_absolute_error',tf.keras.metrics.RootMeanSquaredError()])
```

```python
raw_data1=raw_data[['Pressure','Temperature','DO_11ppg']]
raw_data1
```

```python
X=raw_data1.drop('DO_11ppg',axis=1)
y=raw_data1['DO_11ppg']
X,y=np.array(X),np.array(y)
X.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)
scaler=MinMaxScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
model.fit(X_train,y_train,epochs=1000)
```

```python
model.fit(X_train,y_train,epochs=1000)
```
